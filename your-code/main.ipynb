#### Question 1: Find the questions and answers of last month.

```python
# your code here

# API Query
SITE = StackAPI('stackoverflow')

questions = SITE.fetch('questions', fromdate=1602175166, todate=1604853613,
    min=20)

# loop to get all the questions
questions_list = [q["title"] for q in questions["items"]]

# printing list with the questions
questions_list
```

```python
# API Query
SITE = StackAPI('stackoverflow')

answers = SITE.fetch('answers', fromdate=1602175166, todate=1604853613,
    min=20) #filter='withbody' 

# loop to get all the answers
answers_list = [a["answer_id"] for a in answers["items"]]

# it returns the answers ids instead of the whole answer as it takes too long to load, but the #filter="withbody" would return this info
answers_list
```

#### Question 2: Find the most voted question from last month till today, with at least a score of 5 and tagged with 'python'. 

```python
# your code here

# API Query

SITE = StackAPI('stackoverflow')

voted_questions = SITE.fetch("questions", fromdate=1602175166, todate=1604853613, tagged='python', sort='votes')

# dict keys
voted_questions.keys()
```

```python
voted_questions
```

```python
# loop to get all the questions from last month with at least of score of 5
most_voted = []

for v in voted_questions["items"]:
    if v["score"] > 5:
        most_voted.append(v["title"])
        most_voted.append(v["score"])
```

```python
most_voted
```
